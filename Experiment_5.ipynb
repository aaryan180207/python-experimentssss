**Python Dictionary**

```python
#Creation of Dictionary
thisdict = {
     "brand" : "ford",
     "model" : "mustang",
     "year" : 1964

 }
print(thisdict)
```

```python
# Dictionary takes latest value given for key. It will not allow duplicates
thisdict ={
    "brand":"Ford",
    "model" : "mustang",
    "year": 1964
}
print(thisdict)
```

```python
# Changing/adding key;value in dictionary
car = {
    "brand" : "ford",
    "model" : "mustang",
    "year" : 1964
}
car["color"] = "red"
x = car.values()
print(x)
```
